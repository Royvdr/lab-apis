# 1. Read your exported json file from the data folder

```python
import pandas as pd

df_coffee = pd.read_json('data/coffe_shops.json')

df_coffee.head()
```

# 2. Plot a map with the coffee shops

```python
import folium
from IPython.display import display, HTML

print(f"Total coffee shops in data: {len(df_coffee)}")
if len(df_coffee) > 0:
    print(f"First shop coordinates: Lat {df_coffee['lat'].iloc[0]}, Long {df_coffee['long'].iloc[0]}")
else:
    print("Uh oh... the dataframe is empty!")


df_clean = df_coffee.dropna(subset=['lat', 'long'])

center_lat = df_clean['lat'].mean()
center_lon = df_clean['long'].mean()

coffee_map = folium.Map(location=[center_lat, center_lon], zoom_start=14)

for index, row in df_clean.iterrows():
    folium.Marker(
        location=[row['lat'], row['long']],
        tooltip=row['name'],
        icon=folium.Icon(color='blue', icon='coffee', prefix='fa')
    ).add_to(coffee_map)


display(HTML(coffee_map._repr_html_()))
```
