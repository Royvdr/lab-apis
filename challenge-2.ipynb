# 4. Define a function that returns name, lat, long for starbucks OR restaurants AND for any given location

```python
def foursquare_places(venue, coordinates):
    url = "https://places-api.foursquare.com/places/search"
    
    headers = {
        "Accept": "application/json",
        "Authorization": f"Bearer {clean_token}", 
        "X-Places-Api-Version": "2025-06-17"
    }
    
    params = {
        "query": venue,
        "ll": coordinates, 
        "limit": 10
    }
    
    response = requests.get(url, params=params, headers=headers)
    
    if response.status_code != 200:
        print(f"Oops! API Error: {response.status_code}")
        return []
    
    results = response.json().get('results', [])
    
    clean_places = []
    for place in results:
        name = place.get('name')
        
        lat = place.get('latitude')
        lon = place.get('longitude')
        
        clean_places.append({
            "name": name, 
            "lat": lat, 
            "long": lon
        })
        
    return clean_places

test_results = foursquare_places("Starbucks", "51.831,5.263")

print("Function test successful! Here is the data:")
test_results
```

# 5. Export it into data folder as a json: `data/coffe_shops.json`

```python
import os
import json

os.makedirs('data', exist_ok=True)

with open('data/coffe_shops.json', 'w') as file:
    json.dump(test_results, file, indent=4)

print("Data successfully saved to data/coffe_shops.json! ☕")
```
